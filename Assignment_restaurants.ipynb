```python
import pandas as pd
file_path = r"C:\Users\KIIT0001\OneDrive\Desktop\Swiggy dataset\restaurants.csv"
df = pd.read_csv(file_path)
print(df.head())
```

```python
#checking data types:
print("\nDatatype")  
df.dtypes
```

```python
#displaying dataframe information
print("\nDataframe info:")
df.info()
```

```python
#displaying all the statistical data of the dataset
print("\nDataframe descriptions:")
df.describe()
```

```python
#filtering by city: 
print("\nfiltering restaurants in Kolkata")
df[df['city'] == 'Kolkata']
```

```python
#filtering restaurants by rating: 
print("\nrestaurnats with ratings more than 4")
df[df['ratings'] >= 4]
```

```python
#sorting by ratings: 
print("\nsorting restaurants with ratings from highest to lowest")
df.sort_values(by='ratings', ascending=False)
```

```python
#sorting by total reviews:
print("\nsorting restaurants by total reviews") 
df.sort_values(by='total_reviews', ascending=False)
```

```python
#average rating by city: 
print("\ndisplaying average rating by each city")
df.groupby('city')['ratings'].mean()
```

```python
#top 5 restaurants by ratings: 
print("\ndisplaying top 5 restaurants by ratings")
df.nlargest(5, 'ratings')
```

```python
#bottom 5 restaurants by ratings: 
print("\ndisplaying bottom 5 restaurants by ratings")
df.nsmallest(5, 'ratings')
```

```python
#restaurant ratings distribution: 
print("\nDistribution of restaurants by ratings received")
df['ratings'].plot(kind='hist')
```

```python
#restaurant review density scatter plot: 
print("\nRestaurant review")
df.plot.scatter(x='total_reviews', y='ratings')
```
