```python
import pandas as pd
file_path = r"C:\Users\KIIT0001\OneDrive\Desktop\Swiggy dataset\Orders.csv"
df = pd.read_csv(file_path)
print(df.head())
```

```python
#checking data types:
print("\nDatatype")  
df.dtypes
```

```python
#displaying dataframe information
print("\nDataframe info:")
df.info()
```

```python
#displaying all the statistical data of the dataset
print("\nDataframe descriptions:")
df.describe()
```

```python
#filtering by order status: 
print("\ndisplaying all the cancelled orders")
df[df['order_status'] == 'Cancelled']
```

```python
#sorting by order time
print("\nsorting data by order time")
df.sort_values(by='order_time')
```

```python
#filtering by dish: 
print("\nfiltering data by a particular dish")
df[df['dish_ordered'] == 'Aloo Paratha']
```

```python
#sorting orders by price in descending order
print("\nfiltering orders by price in descending order")
df.sort_values(by='price', ascending=False)
```

```python
#total revenue by dish
print("\ndisplaying total revenue by each dish")
df.groupby('dish_ordered')['price'].sum()
```

```python
#total orders by customer ID 
print("\ndisplaying total orders placed by each customers")
df.groupby('customer_id')['order_id'].count()
```

```python
#count of orders by order status
print("\ndisplaying number of orders by order status")
df['order_status'].value_counts()
```

```python
#top 5 dishes by quantity sold
print("\ndisplaying top 5 orders sold by quantity sold in visual format")
df['dish_ordered'].value_counts().nlargest(5).plot(kind='bar')
```

```python
#order frequency over time: 
df['order_time'].dt.date.value_counts().plot(kind='bar')
```

```python
#delivery partner performance: 
df.groupby('delivery_partner_id')['delivery_time'].mean().plot(kind='bar')
```
